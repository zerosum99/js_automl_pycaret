## 데이터 읽어오기 

```python
path_csv = "/Users/a06411/Documents/GitHub/js_automl_pycaret/lending_club_data/lending_club_2012.csv"
```

```python
accept = pd.read_csv(path_csv)
```

```python
accept.head()
```

```python
accept.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
accept.head()
```

```python
def put(df_csv, label_info, db):
  tblname = label_info["Label"].unique()
  for i in tblname:
    cols = list(label_info[label_info['Label'] == i]['LoanStatNew'].values)    # label별 데이터 분리
    df = df_csv[["id"]+ cols]  #id값 포함
    df.to_sql(i, db, if_exists='replace', index=False)
  db.close()
```

```python
path_db = "/Users/a06411/Documents/GitHub/js_automl_pycaret/lending_club_data/sql_db/lending_club_2012.db"
```

```python
con = sqlite3.connect(path_db)
put(accept,label, con)
```
