## 4. 특정 문자가 들어간 칼럼 조회하기 

## 문자열 자료형에는 str 접근 객체 제공

-  특징은 str 클래스의 메서드 등을 다 제공한다. 

```python
df.columns.str.contains('ty')
```

## 함수를 정의해서 특정 단어만 포함한 칼럼을 조회

```python
self = {}
```

```python
self["_df"] = df
```

```python
self["_df"].columns[self["_df"].columns.str.contains("ty")]
```

```python
class SelectCols :
    def __init__(self,df) :
        self._df = df
                           
    def selectCols(self,exer) :
        return self._df[self._df.columns[self._df.columns.str.contains(exer)]] 

    def select_Cols(self,exer) :
        return self._df[getattr(self,"_df").columns[getattr(self,"_df").columns.str.contains(exer)]]
```

```python
sc = SelectCols(df)
```

## 속성으로 접근해서 처리하기 

```python
sc.selectCols("er")
```

## 문자열로 처리해서 접근하기

```python
sc.select_Cols("er")
```

```python
def selectCols(df,exer) :
    return df[df.columns[df.columns.str.contains(exer)]]
```
