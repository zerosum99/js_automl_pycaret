## load dataset

```python
from pycaret.datasets import get_data 
boston = get_data('boston') 
```

##  init setup

```python
from pycaret.regression import * 
reg1 = setup(boston, target = 'medv')
```

## train model

```python
dt = create_model('dt')
```

# tune model sklearn

```python
tune_model(dt)
```

```python
# tune model optuna
tune_model(dt, search_library = 'optuna')
```
